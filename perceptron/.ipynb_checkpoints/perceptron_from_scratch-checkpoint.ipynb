The perceptron is a simple machine learning algorithm and one of the earliest neural network architectures (Perceptron is a **sigle layer neural network**).

A perceptron represents a **binary linear classifier** that maps a set of training examples (of $d$ dimensional input vectors) onto binary output values using a $d-1$ dimensional hyperplane.

## Training steps

**Step 0**: Initialize the weight vector $w$ and bias $b$ with zeros (or small random values)

**Step 1**: Compute a linear combination of the input features and weights. This can be done in one step for all training examples, using vectorization and broadcasting: $a = X \cdot \omega + b$, where $X$ is a matrix of shape $(n_sample, n_features)$ that holds all training examples, and $\cdot$ denotes the dot product.

**Step 2**: Apply the Heaviside function, which returns binary values:
$$
\hat{y}^{(i)}=1 \text { if } a^{(i)} \geq 0, \text { else } 0
$$

**Step 3**: Compute the weight updates using the perceptron learning rule:
$$
\begin{gathered}
\Delta \boldsymbol{w}=\eta \boldsymbol{X}^{T} \cdot(\hat{\boldsymbol{y}}-\boldsymbol{y}) \\
\Delta b=\eta(\hat{\boldsymbol{y}}-\boldsymbol{y})
\end{gathered}
$$

**Step 4**: Update the weight and bias:
$$
\begin{gathered}
\boldsymbol{w}=\boldsymbol{w}+\Delta \boldsymbol{w} \\
b=b+\Delta b
\end{gathered}
$$

## Coding up a perceptron

```python
# Importing Libraries
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Complete the `Perceptron` class.**

```python
class Perceptron:
    def __init__(self, epochs=3):
        """
        Args:
            epochs: number of training epochs
        """
        self.epochs = epochs
        self.misclassifications = 0
        
        ## Add any other variables you need here
        ## YOUR CODE HERE
        
        self.learning_rate = 0.0015
        self.weights = np.ones((2,1))
        self.bias = 0 # Add bias here in case overfitting since there just 10 dp in this case
    
    def update_weights(self, features, true_labels):
        """
        The weight update rule. Iterates over each weight and updates it.
        Increments self.misclassifications by 1 if there is a misclassification.

        Args:
            features: Dependent variables (x)
            true_label: Target variable (y)
        """
        ## YOUR CODE HERE
        
        
        ## compute the activation
        
        y_predict = self.predict(features)
        
       
        ## compute misclassifications
        
        self.misclassifications = 0
        for i in range(len(features)):
            if y_predict[i][0] != true_labels[i]:
                self.misclassifications += 1
            
            
        ## cumpute weight update
       
        delta_w = self.learning_rate * np.dot(features.T, (true_labels.reshape(len(true_labels),1) - y_predict))
        delta_b = self.learning_rate * np.sum(true_labels - y_predict)
        
        ## update param
        self.weights += delta_w
        self.bias += delta_b
        
        #return self.weights, self.bias
            
        
        
    
    def train(self, features, true_labels, plotting=True):
        """
        features: dependent variables (x)
        true_labels: target variables (y)
        plotting: plot the decision boundary (True by default)
        """
        # Initialize the weights
        ## YOUR CODE HERE
        
        self.misclassifications = 0
        

        # For each epoch
        for epoch in range(self.epochs):
            # Iterate over the training data
            for i in range(len(features)): 
                
                
                if plotting:
                    print("Iteration {}, Misclassifications = {}".
                      format(epoch * len(features) + i+1, self.misclassifications))
                    self.plot_classifier(features, true_labels, features[i])
                
                # Update the weights
                ## YOUR CODE HERE
                
                
                self.update_weights(features, true_labels)
                
                
            print("="*25)
            print("Epoch {}, Accuracy = {}".format(epoch + 1, 1 - self.misclassifications/len(features)))
            print("="*25)
            self.misclassifications = 0
            
    def predict(self, features):
        """
        Predict the label using self.weights.
        
        Args:
            features: dependent variables (x)
            
        Returns:
            The predicted label.
        """
        ## YOUR CODE HERE
         
        ## compute the activation
        
        hyper_plane = np.dot(features, self.weights) + self.bias
        return np.array([1 if elem >= 0 else -1 for elem in hyper_plane])[:, np.newaxis]
        
    
    def plot_classifier(self, features, true_labels, data_point):
        """
        Plot the decision boundary.
        
        Args:
            features: dependent variables (x)
            true_labels: target variables (y)
            data_point: the current data point under consideration
        """
        # Create a mesh to plot
        x1_min, x1_max = features[:, 0].min() - 2, features[:, 0].max() + 2
        x2_min, x2_max = features[:, 1].min() - 2, features[:, 1].max() + 2
        x1x1, x2x2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02),
                             np.arange(x2_min, x2_max, 0.02))
        
        Z = np.zeros(x1x1.shape)
        fig, ax = plt.subplots()
        for i in range(len(x1x1)):
            for j in range(len(x1x1[0])):
                Z[i,j] = self.predict([x1x1[i,j], x2x2[i,j]])

        # Put the result into a color plot
        ax.contourf(x1x1, x2x2, Z, cmap='bwr', alpha=0.3)

        # Plot the training points
        plt.scatter(features[:, 0], features[:, 1], c=true_labels, cmap='bwr')
        #plt.scatter(features[:, 0], features[:, 1], c=features[:, -1], cmap='bwr')
        plt.plot(data_point[0], data_point[1], color='k', marker='x', markersize=12)

        ax.set_title('Perceptron')

        plt.show()
```
